# Landmark .parquet File Exploration
This notebook loads and visualizes a sample sign language landmark .parquet file.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob

train_data_path = "../dataset/train_landmark_files"
```

## 1. Locate a Sample .parquet File

```python
DATA_DIR = train_data_path  # Adjust path as needed
sign_folders = [d for d in os.listdir(DATA_DIR) if os.path.isdir(os.path.join(DATA_DIR, d))]
print('Sign folders:', sign_folders)
sample_sign = sign_folders[0] if sign_folders else None
sample_dir = os.path.join(DATA_DIR, sample_sign) if sample_sign else None
sample_files = glob.glob(os.path.join(sample_dir, '*.parquet')) if sample_dir else []
sample_file = sample_files[50] if sample_files else None
print('Sample file:', sample_file)
```

## 2. Load and Inspect the .parquet File

```python
# if sample_file:
#     df = pd.read_parquet(sample_file)
#     print(df.tail(15))
#     print('Columns:', df.columns)
#     print('Shape:', df.shape)
# else:
#     print('No sample .parquet file found.')

df = pd.read_parquet("../dataset/train_landmark_files/25571/100387913.parquet")
print(df.tail(700))
print('Columns:', df.columns)
print('Shape:', df.shape)
```

## 3. Visualize Landmarks (x, y, z) Over Time

```python
if sample_file:
    plt.figure(figsize=(12,4))
    plt.plot(df['x'], label='x')
    plt.plot(df['y'], label='y')
    plt.plot(df['z'], label='z')
    plt.title('Landmark Coordinates Over Time')
    plt.xlabel('Frame')
    plt.ylabel('Coordinate Value')
    plt.legend()
    plt.show()
```
